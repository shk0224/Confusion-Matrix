# Now seperating x and y:

```python
df.head(1)
```

```python
x=df.drop('DEATH_EVENT', axis=1)
x.head(1)
```

```python
y=df['DEATH_EVENT']
y.head(1)
```

# Split the dataset into x and y

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3, random_state=1)
```

```python
xtrain.shape       #Looking at the shape
ytrain.shape
```

```python
xtest.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc=DecisionTreeClassifier()  #Creating an object for decison tree
```

```python
dtc.fit(xtrain,ytrain)
```

```python
from sklearn import tree
plt.figure(figsize=(20,20))
tree.plot_tree(dtc ,filled=True, rounded=True, feature_names=x.columns)
```

```python
pred=dtc.predict(xtest)
pred
```

```python
ytest
np.array(ytest)
```

# Confusion Matrix

```python
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix
```

```python
classification_report(ytest,pred)
print(classification_report(ytest,pred))
```

```python
cm=confusion_matrix(ytest,pred)
cm
```

```python
sns.heatmap(cm,annot=True)
plt.title('Confusion Matrix')
plt.xlabel('Predicted values')
plt.ylabel('actual values')
```

```python
tp,fn,fp,tn=confusion_matrix(ytest,pred).reshape(-1)
#tp=true positive, fn=false negative, fp=false positice, tn=true negative
```

```python
#Accuracy:
ac= (tp+tn)/(tp+fn+fp+tn)
ac
```

```python
#Precission or positive predictive value:
ppv=tp/(tp+fp)
ppv
```

```python
#Sensitivity/Reccall/Hit Rate/True Positive rate:
TPR=tp/(tp+fn)
TPR
```

```python
#False Positive Rate:
FPR=fp/(fp+tn)
FPR
```

```python
#F1 Measure=(precision+Recall/2)
F1M=(ppv+TPR)/2
F1M
```

```python
#Harmonic mean/F1 Score=(2*Precission*Recall)/(Precission+Recall)
F1= (2*ppv*TPR)/(ppv+TPR)
F1
```

```python
#Specificity/Selectivity/True Negative Rate(TNR)=(True negative)/(true negative+false positive):
TNR=tn/(tn+fp)
TNR
```

```python
#Theta score/critical success index(CSI)=
CSI=(tp)/(tp+fn+fp)
CSI
```

```python
#False Discovery rate:
FDR=fp/(tp+fp)
FDR
```
